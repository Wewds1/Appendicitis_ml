```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
import os

```

```python
df = pd.read_csv('appendicitis_comprehensive_dataset.csv')
```

DATA EXPLORATION

```python
print("DataFrame loaded successfully")
print(df.head())
print(df.info())
print(df.describe())
print("DataFrame info and description printed successfully")
```

```python
print(df.isnull().sum())

categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols)

unique_val = {col: df[col].unique() for col in categorical_cols}
print("Unique values in categorical columns:", unique_val)

numerical_cols = df.select_dtypes(include=[np.number]).columns
print("Numerical columns:", numerical_cols)
```

```python
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
```

```python
plot_cols = ['Age', 'BMI', 'Duration_of_Symptoms_Hours', 'Fever_Temp_C', 'WBC_Count_k_uL', 'Neutrophil_Percentage', 'CRP_Level_mg_L']
plt.figure(figsize=(21, 14))
for i, col in enumerate(plot_cols,1):
    plt.subplot(3, 4, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
```
